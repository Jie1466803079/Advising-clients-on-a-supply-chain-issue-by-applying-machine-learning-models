````python
import seaborn as sns
````

To analyse the dataset, below are snippets of code that you can use as helper functions to visualise different columns within the dataset. They include:

- plot_continuous_distribution = this is to visualise the distribution of numeric columns
- get_unique_values = this is to show how many unique values are present within a column
- plot_categorical_distribution = this is to visualise the distribution of categorical columns
- correlation_plot = this is to plot the correlations between the numeric columns within the data

````python
def plot_continuous_distribution(data: pd.DataFrame = None, column: str = None, height: int = 8):
  _ = sns.displot(data, x='unit_price', kde=True, height=height, aspect=height/5).set(title=f'Distribution of unit price');

plot_continuous_distribution(df)
````

The graph indicates that the majority of products have a unit price below $10.

````python
def plot_continuous_distribution(data: pd.DataFrame = None, column: str = None, height: int = 8):
  _ = sns.displot(data, x='total', kde=True, height=height, aspect=height/5).set(title=f'total sales');

plot_continuous_distribution(df)
````

The graph above shows that the majority of transactions have a total sales below $30.

````python
def correlation_plot(data: pd.DataFrame = None):
  corr = data.corr()
  return corr.style.background_gradient(cmap='coolwarm')

correlation_plot(df)
````

The graph shows that there is a correlation between unit price/quantity and transaction amount. The correlation between unit price and transaction amount is stronger than that between quantity and transaction amount.
